## 컨테이너 타입

- 컨테이너 타입은 여러 개의 값을 다루는 자료형입니다. **list, tuple, set, dictionary**가 있음
  
    > **리스트(list)** : 데이터를 연속적으로 관리, 값을 바꿀 수 있음

    > **튜플(tuple)** : 데이터를 연속적으로 관리, 값을 바꿀 수 없음

    > **세트(set)** : 집합과 같은 속성

    > **딕셔너리(dictionary)** : key-value 쌍

## List 리스트
- 여러 개의 데이터를 묶어서 처리할 때 사용, 인덱스를 통해 데이터를 엑세스함

## 1. List 생성

- 대괄호 [ ] 사용, 각각의 데이터는 콤마로 구분
- [ ] 입력 시 빈 리스트로 생성
- List 의 데이터는 다양하게 저장 가능

```python
score = [80, 90, 100]
score
```

```python
names = ['진진', '뇽', '누리']
names
```

```python
type(score)
type(names)
```

```python
print(type(score), type(names))
```

## 2. List에서 아이템 지정 방법
    - 리스트명 [ 인덱스 ]

### (1) 아이템 조회

```python
score[0]
score[2]
```

```python
len(score)
```

```python
score[-3]
```

```python
print(names[1], names[2], names[-3])
```

### (2) 아이템 변경
    - 리스트명 [ 인덱스 ] = new item

```python
names[0]
```

```python
names[0] = '짭진진'
names[0]
```

```python
names
```

```python
class1 = ['누리', 80, 90, 100, '뇽', 90, 20, 30, False]
class1
```

## 3. List 연산

- 더하기 ( + ) : 두 개의 list 연결
- 곱하기 ( x ) : 곱한 수만큼 반복
- 범위 지정 (클론)
    - 리스트 [ start : end ]
    - 리스트 [ start : end : step ]

- 아이템 제거 : del 리스트 [ 인덱스 ]
- 리스트 조회 : 아이템 항목의 유무 확인 True / False 출력
    - 아이템 in 리스트 객체

```python
list1 = [1, 2, 3, 4]
list2 = [2, 3, 4, 5]
list1 + list2
```

```python
list1 * 3
```

```python
(list1 * 3) + list2
```

```python
list3 = (list1 * 2) + list2
list3
```

```python
list3[0:4]
```

```python
list3[5:len(list3)]
```

```python
list3[::]
```

```python
list3[::2]
```

```python
del list3[0]
list3
```

```python
del list3[0:4]
list3
```

```python
9 in list3
```

## 4. List의 메소드
    - 리스트.메소드( )

### (1) append
    - 리스트의 마지막에 새로운 항목 추가

```python
fruits = ['lemon', 'blueberry']
fruits
```

```python
fruits.append('melon')
fruits
```

### (2) insert
    - 특정 위치에 아이템 추가 : insert ( 인덱스, 추가할 항목 )

```python
fruits.insert(0, 'peach')
fruits
```

### (3) extend 
    - 여러 개의 항목을 마지막에 추가 : extend ( [ 아이템1, 아이템2 .. , 추가 항목 ] )

```python
fruits.extend([0, 2, '딸기'])
fruits
```

### (4) remove
    - 항목 삭제하기 : remove ( 아이템 )

```python
fruits.remove('딸기')
fruits
```

### (5) pop
    - 마지막 아이템 삭제 후 return

```python
fruits.pop()
```

```python
fruits
```

### (6) index
    - 아이템 현재 인덱스 확인 : index ( 아이템 )

```python
fruits.index('lemon')
```

### (7) count
    - 아이템 전체 갯수 확인 : count ( 아이템 )

```python
list4 = (list1 * 2) + list3
list4
```

```python
list4.count(3)
```

### (8) sort
    - 오름차순 정렬 : sort ( )

```python
list4.sort()
list4
```

### (9) reverse
    - 내림차순 정렬 : reverse ( )

```python
list4.reverse()
list4
```
