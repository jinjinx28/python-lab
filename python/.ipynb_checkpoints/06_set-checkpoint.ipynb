- 데이터에 순서가 없고 중복해서 사용이 불가
- 교집합, 합집합, 차집합

### 1. 세트 생성

## 2. Set ( 집합 ) 연산

- 교집합
      > set 객체.intersection ( set 객체 )
  
- 합집합
      > set 객체.union ( set 객체 )

- 차집합
      > set 객체.difference ( set 객체 )

```python
# 교집합
set1 = {1, 2, 3}
set2 = {2, 3, 4}
print (set1, set2)
print(set1.intersection(set2))
```

```python
set1 & set2
```

```python
# 합집합
print(set1.union(set2), set1 | set2)
```

```python
# 차집합
print(set1.difference(set2), set2.difference(set1))
```

## 4. 아이템 추가 / 업뎃
- set 객체.add( 아이템 )
- set 객체.update( set 객체 )

```python
fruits_set.add('bubu')
fruits_set
```

```python
vegetable_set = {'carrot', 'onion'}
vegetable_set
```

```python
fruits_set.update(vegetable_set)
fruits_set
```

## 5. 아이템 삭제
- set 객체.remove( 아이템 ) : 아이템 삭제
- set 객체.clear( ) : 아이템 전체 삭제

```python
fruits_set.remove('bubu')
```

```python
fruits_set
```

```python
fruits_set.clear()
```

## 6. 리스트, 튜플, 세트 간 변환 타입
- list( ), tuple( ), set( )

```python
a = [1, 2, 3, 4, 5]
b = (1, 2, 3, 4, 5)
c = {1, 2, 3, 4, 5}
```

```python
print(type(a), type(b), type(c))
```

```python
a_to_tuple = tuple(a)
a_to_tuple
```

```python
b_to_set = set(b)
b_to_set.remove(4)
b_to_set
```

```python
c_to_list = list(c)
c_to_list.append(5)
c_to_list
```

```python
c_to_set = set(c_to_list)
c_to_list
```
