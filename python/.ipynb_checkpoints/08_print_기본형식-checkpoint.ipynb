## Print ( )

## 1. 기본 문법
- print ( value, ..., sep = '구분기호', end = '\n', flush = False)

```python
a = 5
a
```

```python
b = 3
b
```

```python
a + b
```

```python
print(a)
print(b)
print(a + b)
```

```python
print(1, 2, 3, sep = '-', flush = True)
```

```python
print(a, end = '\t')
print(b)
print(a + b)
```

```python
name = '진진'
age = 23
```

```python
# 1. 콤마로 구분 가능
print('이름 : ', name, ', 나이 : ', age) 
```

```python
# 2. 접합연산자 + 사용
print('이름 : ' + name, ', 나이 : ' + str(age))     # int = 숫자, str = 문자열
```

```python
# 3.% 포맷 (old ver.)
print("이름 : %s , 나이 : %d" % (name, age))
```

```python
# 4. format()
print('이름 : {}, 나이 : {}'.format(name, age))
```

```python
# 5. f-string 
print(f'이름 : {name}, 나이: {age}')
```

```python
# 여러 값을 출력 콤마 구분
print(1, 2, 3)
print(1, 2, 3, sep = ',')
```

```python
# 줄바꿈 없이 옆으로 출력
for i in range(5) :
    print(i)
```

```python
# 줄바꿈 없이 옆으로 출력
lst = [1, 2, 3, 4, 5]
for i in lst:   
    print(i, end='\t')
```

```python
# 빈 줄 출력
print()
```

```python
# * 문자열을 50번 출력
print('*' * 50)
```
