## 반복문
- 반복적인 작업을 수행하는 구문
- for문, while문

### 1. for문

```python
for number in [1, 2, 3, 4, 5]:
    print(number, end='\t')
```

```python
fruits = ['apple','lemon','mango']
for fruit in fruits:
    print(fruit, end='\t')
```

```python
for x in range(0,5,1):  
    print(x)
```

```python
for x in range(5): 
    print(x+1, end=" ")
```

### range ( start, stop, step )
- start : 범위의 시작점, stop : 범위의 끝 지점, step : 증감의 크기
- step = 1인 경우 생략 가능
- step =1, start =0인 경우 start 역시 생략 가능

```python
list1 = list(range(1, 11))
list1
```

```python
print(list(range(10)))
```

```python
print(set(range(10)))
print(tuple(range(10)))
```

```python
for i, fruit in enumerate(fruits):
    print(i, fruit)
```

```python
for i, num in enumerate(range(101, 111)):
    print(i, num, sep=":")
```

```python
fruits2 = { "apple":"🍎", "lemon":"🍋", "tomato":"🍅"}
for f in fruits2:
    print(f)
```

```python
fruits2 = { "apple":"🍎", "lemon":"🍋", "tomato":"🍅"}
for key, emoji in fruits2.items():
    print(key, emoji, sep=",")
```

```python
fruits2 = { "apple":"🍎", "lemon":"🍋", "tomato":"🍅"}
for key, emoji in fruits2.items():
    print(f"과일명:{key}, 이모지:{emoji}")
```

### 2. 중첩 for문

```python
n_list1 = [10, 20]
n_list2 = [30, 40]
for i1, x in enumerate(n_list1):
    for i2, y in enumerate(n_list2):
        print(f"({i1}, {i2}),  {x + y}")
```

```python
dan = 3
for n in range(1,10):
    print(f"{dan} ✖ {n} = {dan * n}")
```

```python
for n in range(1, 10):
    for dan in range(2, 10):
        print(f"{dan} ✖ {n} = {dan*n}", end="\t")
    print()
```

### 3. while문
- 조건에 따라 반복 여부 결정

```python
num = 1
while num <=5:
    print(f"number={num}")
    num += 1
```

```python
k = 0
while True:
    k += 1
    if k > 3:
        break
    else:
        print(k)
```

```python
for n in range(1, 6):
    if n == 2:
        continue
    print(n)
```

### 4. comprehension

```python
numbers = [1, 2, 3, 4, 5]
numbers2 = [k **2 for k in numbers]
print(numbers, numbers2)
```
