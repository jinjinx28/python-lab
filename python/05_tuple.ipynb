## Tuple ( 튜플 )

- 리스트와 유사
- 한 번 입력 한 후에 항목을 변경할 수 없는 특징
- 소괄호()를 사용해서 생성 또는 괄호 없이 생성
- 튜플의 항목은 숫자, 문자열, 불, 리스트, 튜플, 세트, 딕셔너리 등으로 만들 수 있음

## 1. 튜플 생성

```python
tuple1 = (1, 2, 3)
tuple1
```

```python
tuple1[0]
```

```python
tuple1[0:2]
```

## 2. 튜플 아이템 생성

```python
fruits_tuple = ('lemon', 'melon');
fruits_tuple
```

```python
type(fruits_tuple)
```

```python
fruits_tuple[1]
```

```python
fruits_tuple[1] = 'Mango' # 아이템 변경 불가 X
```

## 3. 튜플 아이템 추가 / 삭제 - 실행 불가 X

```python
fruits_tuple
```

```python
fruits_tuple.append('Lemon') # 새로운 아이템 추가 안됨 X
```

```python
fruits_tuple.remove('Apple') # 아이템 삭제 안됨 X
```

## 4. 튜플을 리스트로 형 변환

```python
type(fruits_tuple)
```

```python
fruits_list = list(fruits_tuple)
```

```python
type(fruits_list)
```

```python
fruits_list.append('orange')
fruits_list
```

```python
fruits_tuple = tuple(fruits_list)
```

```python
fruits_tuple
```

```python
f_tuple = ('lemon', 'melon')
f_list = list(f_tuple)
f_list[f_list.index('lemon')] = 'blueberry'
f_tuple = tuple(f_list)
```

```python
print(type(f_tuple), f_tuple)
```

## 5. 튜플 삭제

```python
del fruits_tuple
```
