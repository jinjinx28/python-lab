## 인덱싱 ( indexing )

- ndrray에서 데이터를 검색하는 방법
- 인덱싱 방법
    | 종류 | 방법 | 특징 |
    |------|------|------|
    |단일값 인덱싱|arr[2] | 특정 위치 단일값 출력|
  |슬라이싱|arr[1:4]| 연속된 범위|
  |팬시 인덱싱 |arr[0, 2, 4] | 불연속 위치 다중값 출력|
  |불린 인덱싱 | arr [arr > 5] | 조건에 맞는 값 출력 |

### (1) 단일값 인덱싱

```python
import numpy as np
```

```python
arr = np.arange(1, 10)
arr
```

```python
arr[2]
```

```python
print(type(arr))
print(type(arr[2]))

print(arr.dtype)
print(arr[2].dtype)

print(arr.shape)
print (arr[2].shape)
```

```python
print(arr)
print(arr[0])
print(arr[3])
print(arr[5])
print(arr[-2])
print(arr[-5])
```

### (2) 값 변경

```python
arr[0] = 100
print(arr)
arr[-1] = 900
print(arr)
```

### (3) 2차원 배열
- 2차원 배열은 [행, 열] 기준으로 검색

```python
arr = np.arange(1, 10)
arr
```

```python
arr2 = np.reshape(arr, (3, 3))
arr2
```

```python
print(arr2.shape)
print(arr2[0, 2])
print(arr2[2, 1])
print(arr2[1, 1])
print(arr2[-1, -1])
```

### (4) 슬라이싱
- ' : ' 기호를 이용하여 연속된 데이터를 출력

```python
array1 = np.arange(1, 10)
array1
```

```python
array1[0:3]
```

```python
print(array1[3:6])
print(array1[:3])
print(array1[::2])
print(array1[::-1])
```

#### 2차원 슬라이싱

```python
arr = np.arange(1, 10).reshape((3, 3))
```

```python
print(arr[0:1, 0:3])
print(arr[0:2, 0:2])
print(arr[1:3, 1:3])
print(arr[:2, :3])
```

```python
print(arr[0], arr[0].shape)
print(arr[2], arr[2].shape)
```

### (5) 팬시 인덱싱
- 불연속적인 여러 위치의 값을 인덱싱하여 출력

#### - 1차원 배열의 팬시 인덱싱

```python
array1 = np.arange(1, 10)
array1
```

```python
print(array1[0])
print(array1[0:3])
print(array1[[0, 2, 3]])
```

```python
array1
```

```python
value = array1[4:][[2, 4]]
print(value, value.shape)
```

#### - 2차원 배열 팬시 인덱싱

```python
arr = np.arange(1, 10).reshape((3, 3))
arr
```

```python
print(arr[[0, 1], 2])
print(arr[[0, 2], 2])
print(arr[[0, 2], [1]])
print(arr[[0, 2], 1:2])
```

### (6) 불린 인덱싱
- 조건식을 사용하여 인덱싱
- 데이터 필터링에 자주 사용됨

```python
arr = np.arange(1, 10)
arr
```

```python
arr[arr > 5]
```

```python
arr_bool = arr < 5
arr_bool
```

```python
arr_bool = arr > 5
print(arr)
print(arr_bool, arr[arr_bool], arr[arr > 5])
```

```python
print(arr[arr%2 == 0])
print(arr[arr%2 == 1])
```

```python
print(arr[(arr%2 == 0) | (arr%5 == 0)])
```

```python
arr = np.arange(1, 10)
arr
```

```python
target = []
type(target)
```

```python
target = []
for n in range(1, len(arr) + 1) :
    if n > 5 :
        target.append(n)
print(target, type(target))
arr2 = np.array(target)
print(arr2, type(arr2))
```

```python
list(range(1, len(arr) + 1))
```

```python
arr1 = arr[arr  > 5]
print(arr1, type(arr1))
```

```python
names = np.array(['진진', '뇽', '누리', '가나디', '고냐니'])
scores = np.array([80, 90, 100, 20, 50])

print(names [scores >= 80])

print(scores [scores >= 90])
```
