## if문

- 지정한 조건에 따라 다르게 분기해 명령을 수행
- 조건의 만족 여부에 따라서 코드 수행 결과가 달라짐
- 단일 조건에 따른 분기

< 구문 >
- 코드블럭은 들여쓰기로만 구분
- 보통은 한 줄에 한 문장, 길어지면 "로 연결
- 한 줄에 여러 문장을 쓸 때는 ; 로 구분
- 주석은 #로 시작

```python
x = 10
if x >= 10: 
    print('Pass')
```

```python
### 비교 연산자
1. == : 같다
2. != : 같지 않다
3. < : 작다
4. > : 크다
5. <= : 작거나 같다
6. >= : 크거나 같다
```

### 논리 연산자
: 논리 연산자 의미 활용
    ex ) 설명 and 논리곱 A and B A와 B가 모두 참이면 참이고 그 외에는 거짓 or 논리합 A or B A와 B 중 하나라도 참이면 참이고 둘 다 거짓이면 거짓 not 논리부정 not A A가 참이면 거짓이고 거짓이면 참

```python
today_temp = 25
if today_temp >= 23:
    print('아이스 아메리카노')
```

```python
today_temp = 20
if today_temp >= 23:
    print("아이스 아메리카노")
else:
    print("따뜻한 아메리카노")
```

```python
today_temp = 0
if today_temp >= 23:
    print("아이스 아메리카노")
elif today_temp == 0:
    print("따뜻한 아메리카노")
else:
    print("유자차")
```
