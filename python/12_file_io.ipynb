## file 열기 (open ( ))
- 내장 함수인 open( ) 함수 이용해 파일을 연 다음 Read, Write 가능
- 형식 : file = open ( 파일 주소 )
- 속성 : 
    1. r : read 모드로 파일 열기 ( 기본 )
    2. w : write 모드로 파일 열기
    3. a : append 모드로 파일 열기, 같은 이름의 파일이 없으면 w 모드로 생성됨
    4. b : 바이너리 파일 모드로 열기
    5. t : 텍스트 파일 모드로 열기

### 1. 파일 쓰기
- 파일 쓰기 하려면 open ( ) 함수의 쓰기 모드로 열어야함
- 파일 쓰기가 종료하면 파일을 닫아야함

```python
f = open('myFile.txt', 'w')
f.write('Hello~ myFile~ ^^')
f.close()
```

### 2. 파일 읽기
- 파일 읽기를 하려면, open ( ) 함수의 읽기( r ) 모드로 열어야 함
- 파일 읽기가 종료하면 파일을 닫아야함

```python
f = open('myFile.txt', 'r')
f_text = f.read()
f.close()

f_text
```

```python
# myFile_ko.txt 파일 생성(인코딩)
f = open('myFile_ko.txt', 'w', encoding='cp949')
f.write('안녕~ 파이썬 ^^')
f.close()

# myFile_ko.txt 파일 읽기
f = open('myFile_ko.txt', 'r')
# print(f.read())
text = f.read()
f.close()

text
```

### 3. 파일에 문자열 한 줄씩 쓰기

```python
f = open('myFile_multi.txt', 'w')
for num in range(1, 6):
    str = f" num:{num}\n" 
    f.write(str)
f.close()
```

### 4. 파일을 한 줄씩 읽기
- readline ( )
- readlines ( ) : 파일 전체를 한줄씩 읽어 리스트 객체로 반환

```python
f = open('myfile_multi.txt', 'r')
line1 = f.readline()
f.close()
print(line1)
```

```python
f = open('myFile_multi.txt')
line = f.readline() 
while line:
    print(line, end=" ")
    line = f.readline()
print("-"*10)    
f.close()
```

```python
f = open('myFile_multi.txt', 'r')
lines = f.readlines()
f.close()

type(lines)
```

```python
for line in lines:
    print(line)
```
