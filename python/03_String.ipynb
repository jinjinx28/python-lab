## 2. 문자열 다루기

- 더하기 연산자 (+) / 곱하기 연산자 (*) 이용
    - (+) 연산자 : 문자열 결합 통해 하나로 생성함
    - (*) 연산자 : 곱한 수만큼 문자열 반복 

```python
a = "Hello "
b = "I'm nuri "
c = a + b
print(c)
```

```python
a * 3 + b * 3
```

```python
(a + b) * 3
```

```python
print(a)
print(b)
print(c)
```
