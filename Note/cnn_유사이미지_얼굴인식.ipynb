```python
import numpy as np
from PIL import Image
import glob
```

#### 유사 이미지

```python
dirNamess = ['Aiden', 'Andrew', 'Cathy']
```

```python
for name in dirNamess :
    count = 0
    for file in sorted(glob.glob(f"../Data/Face/{name}/*.jpg")) :
        img = Image.open(file)
        for angle in range(-15, 15, 1) :
            count += 1
            # 회전
            img2 = img.rotate(angle)
            img2.save(f"../Data/FaceGray/{name}/rotation_{count}.jpg")
```
