```python
import numpy as np
from PIL import Image
import glob
```

#### мң мӮ¬ мқҙлҜём§Җ

```python
dirNames = ['Aiden', 'Andrew', 'Cathy']
```

```python
for name in dirNames :
    count = 0
    for file in sorted(glob.glob(f"../Data/Face/{name}/*.jpg")) :
        img = Image.open(file)
        for angle in range(-15, 15, 1) :
            count += 1

            # нҡҢм „
            img2 = img.rotate(angle)
            img2.save(f"../Data/FaceGray/{name}/rotation_{count:04}.jpg")

            # н”ҢлҰҪ
            img3 = img2.transpose(Image.FLIP_LEFT_RIGHT)
            img3.save(f"../Data/FaceGray/{name}/flip_{count:04}.jpg")
```

```python
# мқҙлҰ„лі„ м „мІҙ file к°ҜмҲҳ
18 * 30 * 2
```

#### м „мІҙ мӮ¬м§„ мӨ‘ мөңлҢҖ н•ҙмғҒлҸ„ м°ҫкё°
 : Featureмқҳ мҲҳлҘј лҸҷмқјн•ҳкІҢ н•ҳкё° мң„н•Ё

```python
widthRatio = []
heightRatio = []

count = 0       # нҢҢмқј к°ҜмҲҳ нҷ•мқё

for name in dirNames :
    count = 0 
    for file in sorted(glob.glob(f"../Data/FaceGray/{name}/*.jpg")) :
        img = np.array(Image.open(file), dtype=np.int32)
        widthRatio.append(img.shape[1])
        heightRatio.append(img.shape[0])
        count += 1
    print(f"{name} : {count}")
```

```python
# мөңлҢҖ н•ҙмғҒлҸ„
print("л„Ҳл№„ мөңлҢҖ н•ҙмғҒлҸ„ : ", np.max(widthRatio))
print("лҲӮмқҙ мөңлҢҖ н•ҙмғҒлҸ„ : ", np.max(heightRatio))
```
