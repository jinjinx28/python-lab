#### iris.csv를 학습해 모델을 저장
- svm (svm_iris.h5)
- randomForest (rf_iris.h5)
- knn (knn_iris.h5)

저장된 모델 불러 예측하기
- SepalLength : 5.1
- SepalWidth : 3.0
- PetalLength : 1.3
- PetalWidth : 0.2

```python
import os
os.environ["PYTHONIOENCODING"] = "utf-8"
```

```python
import pandas as pd
import warnings
from sklearn.svm import SVC
from sklearn.calibration import CalibratedClassifierCV  
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
import joblib
```

```python
df_iris = pd.read_csv("../Data/iris.csv")
```

```python
x = df_iris[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
```

```python
y = df_iris['Name']
```

```python

svm_model = CalibratedClassifierCV(SVC(kernel='rbf'), ensemble=False)
rf_model = RandomForestClassifier(n_estimators=100)
knn_model = KNeighborsClassifier(n_neighbors=5)

svm_model.fit(x, y)
rf_model.fit(x, y)
knn_model.fit(x, y);

print("--- 1. 모델 학습 완료 ---")
```

```python
joblib.dump(svm_model, 'svm_iris.h5')
joblib.dump(rf_model, 'rf_iris.h5')
joblib.dump(knn_model, 'knn_iris.h5')
print("--- 2. 모델 저장 완료 (.h5) ---")
```

```python
loaded_svm = joblib.load('svm_iris.h5')
loaded_rf = joblib.load('rf_iris.h5')
loaded_knn = joblib.load('knn_iris.h5')
print("--- 3. 저장된 모델 로드 완료 ---")
```

```python
new_data = pd.DataFrame([{
    'SepalLength': 5.1,
    'SepalWidth': 3.0,
    'PetalLength': 1.3,
    'PetalWidth': 0.2
}])

pred_svm = loaded_svm.predict(new_data)
pred_rf = loaded_rf.predict(new_data)
pred_knn = loaded_knn.predict(new_data)
```

```python
print("\n[4. 최종 예측 결과]")
print(f"SVM 모델 예측 결과           : {pred_svm[0]}")
print(f"Random Forest 모델 예측 결과 : {pred_rf[0]}")
print(f"KNN 모델 예측 결과           : {pred_knn[0]}")
```
